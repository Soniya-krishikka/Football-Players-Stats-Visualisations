```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('C:\\Users\\subal\\OneDrive\\Desktop\\Documents\\Project\\python\\Football Players 2015-2024\\Latest Football  Players 2024 Data.csv')
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for the plots
sns.set(style="whitegrid")

# Visualizing the distribution of matches played
plt.figure(figsize=(12, 6))
sns.histplot(df['Matches'], kde=True, bins=20, color='green')
plt.title('Distribution of Matches Played')
plt.xlabel('Matches')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the distribution of goals scored
plt.figure(figsize=(12, 6))
sns.histplot(df['Goals'], kde=True, bins=20, color='red')
plt.title('Distribution of Goals Scored')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.show() 
```

```python
# Visualizing the distribution of assists made
plt.figure(figsize=(12, 6))
sns.histplot(df['Assists'], kde=True, bins=20, color='blue')
plt.title('Distribution of Assists Made')
plt.xlabel('Assists')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the relationship between Matches and Goals
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Matches', y='Goals', hue='Seasons Ratings', palette='magma')
plt.title('Relationship between Matches and Goals')
plt.xlabel('Matches')
plt.ylabel('Goals')
plt.legend(title='Seasons Ratings')
plt.show()
```

```python
# Visualizing the average seasons ratings for each team
plt.figure(figsize=(20, 14))
team_ratings = df.groupby('Teams')['Seasons Ratings'].mean().sort_values(ascending=False)
sns.barplot(x=team_ratings.values, y=team_ratings.index, palette='magma')
plt.title('Average Seasons Ratings for Each Team')
plt.xlabel('Average Seasons Ratings')
plt.ylabel('Teams')
plt.show()
```

```python
# Identifying top players by season based on their ratings from the original dataframe
top_players_by_season = df.loc[df.groupby('Seasons')['Seasons Ratings'].idxmax()]

# Visualizing the top players by season
plt.figure(figsize=(14, 4))
sns.barplot(data=top_players_by_season, x='Seasons', y='Seasons Ratings', hue='Players', dodge=False, palette='viridis')
plt.title('Top Players by Season')
plt.xlabel('Seasons')
plt.ylabel('Seasons Ratings')
plt.legend(title='Players', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering data for top players Messi and Ronaldo over the seasons
top_players = ['Lionel Messi', 'Cristiano Ronaldo']
top_players_data = df[df['Players'].isin(top_players)]

# Plotting the top players' ratings over time
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_players_data, x='Seasons', y='Seasons Ratings', hue='Players', marker='o', palette='magma')
plt.title('Top Players\' Ratings Over Time')
plt.xlabel('Seasons')
plt.ylabel('Seasons Ratings')
plt.legend(title='Players', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the top players' goals per season
plt.figure(figsize=(18, 6))
sns.lineplot(data=top_players_data, x='Seasons', y='Goals', hue='Players', marker='o', palette='viridis')
plt.title('Top Players\' Goals Per Season')
plt.xlabel('Seasons')
plt.ylabel('Goals')
plt.legend(title='Players', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the top players' assists per season
plt.figure(figsize=(18, 6))
sns.lineplot(data=top_players_data, x='Seasons', y='Assists', hue='Players', marker='o', palette='magma')
plt.title('Top Players\' Assists Per Season')
plt.xlabel('Seasons')
plt.ylabel('Assists')
plt.legend(title='Players', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate goal-to-match ratio for Lionel Messi
messi_data = df[df['Players'] == 'Lionel Messi']
messi_data['Goal-to-Match Ratio'] = messi_data['Goals'] / messi_data['Matches']

# Plotting Lionel Messi's goal-to-match ratio per season
plt.figure(figsize=(18, 6))
sns.lineplot(data=messi_data, x='Seasons', y='Goal-to-Match Ratio', marker='o', color='darkred')
plt.title('Lionel Messi\'s Goal-to-Match Ratio Per Season')
plt.xlabel('Seasons')
plt.ylabel('Goal-to-Match Ratio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering data for Cristiano Ronaldo
ronaldo_data = df[df['Players'] == 'Cristiano Ronaldo']

# Plotting Cristiano Ronaldo's goals per season
plt.figure(figsize=(18, 6))
sns.lineplot(data=ronaldo_data, x='Seasons', y='Goals', marker='o', color='darkblue')
plt.title('Cristiano Ronaldo\'s Goals Per Season')
plt.xlabel('Seasons')
plt.ylabel('Goals')
plt.xticks(rotation=45)
plt.show()
```

```python
df_num = df.select_dtypes(exclude='object')

# Creating a correlation heatmap of features
plt.figure(figsize=(8, 4))
correlation_matrix = df_num.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
# Analyzing player performance by team
team_performance = df.groupby('Teams').agg({
    'Goals': 'sum',
    'Assists': 'sum',
    'Matches': 'sum',
    'Seasons Ratings': 'mean'
}).reset_index()
```

```python
team_performance.head()
```

```python
# Visualizing team performance metrics

# Plotting total goals by team
plt.figure(figsize=(18, 6))
top_teams_goals = team_performance.sort_values(by='Goals', ascending=False).head(15)
sns.barplot(data=top_teams_goals, x='Goals', y='Teams', palette='magma')
plt.title('Top 15 Teams by Total Goals')
plt.xlabel('Total Goals')
plt.ylabel('Teams')
plt.show()
```

```python
# Plotting total assists by team
plt.figure(figsize=(18, 6))
top_teams_assists = team_performance.sort_values(by='Assists', ascending=False).head(15)
sns.barplot(data=top_teams_assists, x='Assists', y='Teams', palette='viridis')
plt.title('Top 15 Teams by Total Assists')
plt.xlabel('Total Assists')
plt.ylabel('Teams')
plt.show()
```

```python
# Plotting average season ratings by team
plt.figure(figsize=(18, 6))
top_teams_ratings = team_performance.sort_values(by='Seasons Ratings', ascending=False).head(15)
sns.barplot(data=top_teams_ratings, x='Seasons Ratings', y='Teams', palette='plasma')
plt.title('Top 15 Teams by Average Season Ratings')
plt.xlabel('Average Season Ratings')
plt.ylabel('Teams')
plt.show()
```

```python
# Aggregating data by teams and seasons to show performance trends over time
team_season_performance = df.groupby(['Teams', 'Seasons']).agg({
    'Goals': 'sum',
    'Assists': 'sum',
    'Matches': 'sum',
    'Seasons Ratings': 'mean'
}).reset_index()
```

```python
team_season_performance.head()
```

```python
# Filtering top teams for visualization
top_teams = team_performance.sort_values(by='Goals', ascending=False).head(5)['Teams']

# Plotting team performance trends over time for the top teams
plt.figure(figsize=(18, 6))
for team in top_teams:
    team_data = team_season_performance[team_season_performance['Teams'] == team]
    sns.lineplot(data=team_data, x='Seasons', y='Goals', label=f'{team} Goals')
    sns.lineplot(data=team_data, x='Seasons', y='Assists', label=f'{team} Assists')

plt.title('Top Teams Performance Trends Over Time')
plt.xlabel('Seasons')
plt.ylabel('Performance Metrics')
plt.legend(title='Teams & Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting average season ratings trends over time for the top teams
plt.figure(figsize=(18, 6))
for team in top_teams:
    team_data = team_season_performance[team_season_performance['Teams'] == team]
    sns.lineplot(data=team_data, x='Seasons', y='Seasons Ratings', label=f'{team} Ratings')

plt.title('Top Teams Average Season Ratings Over Time')
plt.xlabel('Seasons')
plt.ylabel('Average Season Ratings')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate goal-to-match ratio for each team
team_performance['Goal-to-Match Ratio'] = team_performance['Goals'] / team_performance['Matches']

# Plotting goal-to-match ratio across teams
plt.figure(figsize=(18, 6))
top_teams_goal_ratio = team_performance.sort_values(by='Goal-to-Match Ratio', ascending=False).head(15)
sns.barplot(data=top_teams_goal_ratio, x='Goal-to-Match Ratio', y='Teams', palette='viridis')
plt.title('Top 15 Teams by Goal-to-Match Ratio')
plt.xlabel('Goal-to-Match Ratio')
plt.ylabel('Teams')
plt.show()
```
